```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('smartphones.csv')
df
```

```python
df.isna().sum()
```

```python
IPhone = df[df['Brand'].isin(['Apple'])]
# plt.figure(figsize=(14,7))
fig = px.scatter(data_frame=IPhone, x='Model', y='Final Price', color='Storage')
fig.show()
```

```python
unaffordable = df[(df['Final Price'] >= 1000)]

# plt.figure(figsize=(20,8))
fig = px.scatter(data_frame=unaffordable, x='Model',y='Brand' ,color='Final Price')
fig.show()
```

```python
cheapest = df[(df['Final Price'] >= 50)]
fig = px.scatter(data_frame=cheapest, y='Model',x='Brand' ,color='Storage', width=2000, height=1000)
fig.show()
```
