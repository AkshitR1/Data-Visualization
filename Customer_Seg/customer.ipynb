```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('customer_segmentation_data.csv')
df = df.set_index('id')
df.head()
```

```python
df.dtypes
```

```python
sns.pairplot(df, vars=['age', 'income', 'spending_score', 'membership_years', 'purchase_frequency', 'last_purchase_amount'])
plt.show()
```

```python
plt.figure(figsize=(12, 7))

hist, bins, _ = plt.hist(df['spending_score'], bins=np.arange(0, 101, 5), edgecolor='k', alpha=0.7, color='#9284DF')
plt.title("Histogram of Spending score")
plt.xlabel("Score")
plt.ylabel("Count")

plt.grid(True)

plt.xticks(np.arange(0, 101, 5))
plt.xlim(0, 100)

plt.rc('font', family='serif')
plt.legend(['Spend Score'])
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df,x='spending_score',y='income',hue='gender', palette='bright')
plt.show()
```

```python
plt.figure(figsize=(18,6))
sns.lineplot(data=df,x='income',y='purchase_frequency',hue='gender' ,palette='bright')
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.countplot(data = df, x = 'gender', color='#3B0F8E')
plt.show()
```

```python

sns.scatterplot(data=df, x = 'age' , y= 'income', hue='gender', palette='dark')
plt.show()
```
