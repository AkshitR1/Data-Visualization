```python
top_defense = df['GA'].value_counts().head(10)
top_team = df['Squad'].value_counts().head(10)
plt.figure(figsize=(12, 6))

sns.barplot(x=top_defense.values, y=top_team.values, orient="h", palette= 'husl')

plt.title("Top 10 Defense")
plt.xlabel("GA")
plt.ylabel("Squad")
```
