```python
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('universal_top_spotify_songs.csv')

df

df
```

```python
dk = df.drop('spotify_id', axis=1)

dk = dk.set_index('daily_rank')

top50 = dk.head(50)
```

```python
sns.scatterplot(x='loudness', y='popularity', data=top50, hue='tempo', palette='rocket')
plt.title('Loudness vs Popularity  ')
plt.show()
```
