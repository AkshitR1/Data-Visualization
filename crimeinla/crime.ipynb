```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
data = pd.read_csv('data2.csv')

data
```

```python
Crime_type = data["Crm Cd Desc"]

print(Crime_type)
```

```python
AreaName = data["AREA NAME"]

print(AreaName)
```

```python
data = data.set_index('DR_NO')
```

```python
data.tail()
```

```python
data['LOCATION'].fillna('Unknown', inplace=True)
```

```python
df_filtered = data[(data['Vict Age'] >0)& (data['Vict Age'].notna())]
df_filtered.loc[:,'Vict Age'].head(10)
```

```python
plt.figure(figsize=(12, 7))

hist, bins, _ = plt.hist(df_filtered['Vict Age'], bins=np.arange(0, 101, 5), edgecolor='k', alpha=0.7, color='#9284DF')
plt.title("Histogram of Victim Ages for LA 2020-2023")
plt.xlabel("Age")
plt.ylabel("Count")

plt.grid(True)

for i in range(len(bins) - 1):
    plt.text(bins[i] + 2.5, hist[i] + 100, str(int(hist[i])), fontsize=10, ha='center', va='bottom')

plt.xticks(np.arange(0, 101, 5))
plt.xlim(0, 100)

plt.rc('font', family='serif')
plt.legend(['Victim Ages'])
```

```python
victim_gender_counts = df_filtered['Vict Sex'].value_counts()

plt.figure(figsize=(12, 12))
plt.pie(victim_gender_counts, labels=victim_gender_counts.index, autopct='%1.1f%%', startangle=0 )
plt.title("Victim Gender Distribution")
```

```python
descode = ['A - Other Asian B - Black C - Chinese D - Cambodian F - Filipino G - Guamanian H - Hispanic/Latin/Mexican I - American Indian/Alaskan Native J - Japanese K - Korean L - Laotian O - Other P - Pacific Islander S - Samoan U - Hawaiian V - Vietnamese W - White X - Unknown Z - Asian Indian']
plt.figure(figsize=(13, 9))
ax = sns.countplot(data=df_filtered, x='Vict Descent', palette='bright')
plt.xlabel('Victim Descent')
plt.ylabel('Count')
plt.title('Barplot of Victim Descent(Color)')
plt.grid(True)
vict_descent_counts = df_filtered['Vict Descent'].value_counts()
xticks = ax.get_xticks()
for i, p in enumerate(ax.patches):
    ax.text(p.get_x() + p.get_width()/2, p.get_height(), 
            f'{vict_descent_counts.iloc[xticks[i]]}\n', 
            ha='center', va='bottom', size=12)
    

plt.show()
```

```python
top_crime_types = df_filtered['Crm Cd Desc'].value_counts().head(10)

plt.figure(figsize=(12, 6))

sns.barplot(x=top_crime_types.values, y=top_crime_types.index, orient="h", palette= 'husl')

plt.title("Top 10 Crime Types")
plt.xlabel("Count")
plt.ylabel("Crime type")

plt.show()
```

```python
data['Date Rptd'] = pd.to_datetime(data['Date Rptd'], format='%m/%d/%Y %I:%M:%S %p')
crimesover_time = data['Date Rptd'].dt.to_period('M').value_counts().sort_index()
```

```python
plt.figure(figsize=(14, 8))
crimesover_time.plot(kind='line', color='purple')
plt.xlabel('Date')
plt.ylabel('Number of Crimes')
plt.title('Crimes Over Time')
plt.grid(True)
plt.tight_layout()
plt.show()
```
