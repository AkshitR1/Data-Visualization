```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
df = pd.read_csv('NBA Stats 202324 All Stats  NBA Player Props Tool (4).csv')
```

```python
dk  = df.set_index('RANK')

df['TEAM'].unique()
```

```python
beststamina = df[["NAME", "MPG"]].sort_values("MPG", ascending=False)
beststamina.head()
```

```python
bestDefRtg = df[["NAME", "DRtg"]].sort_values("DRtg", ascending=False)
bestDefRtg.head(10)
```

```python
Lakers_players = []
for i, j in df[["NAME", "TEAM"]].values:
    if i not in Lakers_players and j == "Lal":
        Lakers_players.append(i)
        
print(Lakers_players)
```

```python
best_shooter_lakers = {}
for i, j in df[["NAME", "3P%"]].values:
    if i in Lakers_players:
        best_shooter_lakers[i] = j
print(best_shooter_lakers)
# plt.figure(figsize=(14,6))
# plt.bar(best_shooter_lakers.keys(), best_shooter_lakers.values())
# plt.show()
```

```python
Most_turnovers = {}#all the players under 30
for i, j in df[["NAME", "TPG"]].values:
    if j >= 2:
        Most_turnovers[i] = j

plt.figure(figsize = (60,10))
plt.bar(Most_turnovers.keys(), Most_turnovers.values(), color = "salmon")
plt.show()
```

```python
best_scorer = df[[ 'NAME','PPG']].sort_values(by='PPG', ascending=False).head(10)
sns.barplot(x='PPG', y='NAME', data=best_scorer, palette='mako')
plt.title('Top 10 Scorers in the Playoffs')
plt.show()
```

```python
best_passer = df[[ 'NAME','APG']].sort_values(by='APG', ascending=False).head(10)
sns.barplot(x='APG', y='NAME', data=best_passer, palette='mako')
plt.title('Top 10 Passers in the Playoffs')
plt.show()
```

```python
sns.scatterplot(x='AGE', y='USG%', data=df, hue='POS', palette='rocket')
plt.title('Plot AGE vs USAGE per Game')
plt.show()
```

```python
sns.scatterplot(x='TPG', y='APG', data=df, hue='MPG', palette='rocket')
plt.title(' Turnover vs Asists per Game')
plt.show()
```

```python
sns.scatterplot(x='AGE', y='TS%', data=df, hue='PPG', palette='mako')
plt.title(' AGE vs TS% per Game')
plt.show()
```
