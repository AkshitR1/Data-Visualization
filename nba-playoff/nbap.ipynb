```python
Lakers_players = []
for i, j in df[["NAME", "TEAM"]].values:
    if i not in Lakers_players and j == "Lal":
        Lakers_players.append(i)
        
print(Lakers_players)
```

```python
best_shooter_lakers = {}
for i, j in df[["NAME", "3P%"]].values:
    if i in Lakers_players:
        best_shooter_lakers[i] = j
print(best_shooter_lakers)
# plt.figure(figsize=(14,6))
# plt.bar(best_shooter_lakers.keys(), best_shooter_lakers.values())
# plt.show()
```
